```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Preparation
data_path = "https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/4688036a-cd89-45f7-a0c3-a34dbe35a50c/download/confirmed-cases-since-120465.xlsx"
df = pd.read_excel(data_path)

# Step 2: Exploratory Data Analysis
# Display the first few rows of the dataset
print(df.head())

# Check the structure of the dataset
print(df.info())

# Step 3: Data Cleaning
# Handle missing values, if any
df.dropna(inplace=True)
```

```python
# Step 4: Creating a data table for the correlation between sex and the number of infections

plt.figure(figsize=(10, 6))
sex_counts = df["sex"].value_counts()
sns.barplot(x=sex_counts.index, y=sex_counts.values)
plt.xlabel("Sex")
plt.ylabel("Number of Infections")
plt.title("Correlation between Sex and Number of Infections")
plt.show()

sex_counts = df["sex"].value_counts()
correlation_table = pd.DataFrame({"Sex": sex_counts.index, "Number of Infections": sex_counts.values})
print(correlation_table)
```

```python
# Step 4: Creating the bar plot for the correlation between province_of_onset and the number of infections
plt.figure(figsize=(10, 6))
province_of_onset_counts = df["province_of_onset"].value_counts().head(20)
sns.barplot(x=province_of_onset_counts.index, y=province_of_onset_counts.values)
plt.xlabel("Province of Onset")
plt.ylabel("Number of Infections")
plt.title("Top 20 Provinces of Onset with the Highest Number of Infections")
plt.xticks(rotation=45)
plt.show()

province_of_onset_counts = df["province_of_onset"].value_counts().head(20)
correlation_table = pd.DataFrame({"Province of Onset": province_of_onset_counts.index, "Number of Infections": province_of_onset_counts.values})
print(correlation_table)
```

```python
# Step 2: Convert "Notified date" to datetime
df["Notified date"] = pd.to_datetime(df["Notified date"])

# Step 3: Count the number of cases per day
daily_cases = df["Notified date"].value_counts().sort_index()

# Step 4: Plotting the daily inflection over time
plt.figure(figsize=(12, 6))
daily_cases.plot(marker='o')
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.title("Daily Inflection of Cases over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Step 2: Convert "Notified date" to datetime
df["Notified date"] = pd.to_datetime(df["Notified date"])

# Step 3: Count the number of cases per day
daily_cases = df["Notified date"].value_counts().sort_index()

# Step 4: Count the number of cases by sex
sex_counts = df["sex"].value_counts()

# Step 5: Plotting the trend over time using subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot the daily inflection
ax1.plot(daily_cases.index, daily_cases.values, marker='o')
ax1.set_xlabel("Date")
ax1.set_ylabel("Number of Cases")
ax1.set_title("Daily Inflection of Cases over Time")

# Plot the sex distribution
ax2.bar(sex_counts.index, sex_counts.values, color='orange', alpha=0.7)
ax2.set_xlabel("Sex")
ax2.set_ylabel("Number of Cases")
ax2.set_title("Sex Distribution")

plt.tight_layout()
plt.show()
```
